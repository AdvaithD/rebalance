```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
table = pd.read_csv("crypto-markets.csv", parse_dates=True)
table
```

```python
table = table[table["date"] > "2018"]
table
```

```python
table = table[["date", "symbol", "market", "close"]]
table
```

```python
table = table[table["market"] > 0]
table
```

```python
table = table.groupby(["date"]).apply(lambda group: group.sort_values("market", ascending=False).head(20)).set_index(["date", "symbol"], drop=True)
table
```

```python
first = table.loc["2018-01-01", ["market"]]
last  = table.loc["2018-05-19", ["market"]]

merged = pd.concat([first, last], axis=1).fillna(0)
merged.columns = ["1/1/2018", "5/19/2018"]

merged.plot.bar(title="Top 20 Composition by Market Cap: 1/1/18 x 5/19/18", figsize=(18, 8))

plt.ylabel("market cap (billions)")
plt.xlabel("symbol")

plt.show()
```

```python
def allocate(df, cap=0.1):
    df['allocation'] = df['market'] / df['market'].sum()
    
    for index in np.arange(len(df)):
        overflow = df.iloc[index, -1] - cap
        
        if(overflow > 0):
            df.iloc[index, -1] = cap
            rest = df.iloc[index + 1:, -3]
            nested_alloc = rest / rest.sum()
            df.iloc[index + 1:, -1] += overflow * nested_alloc
    
    return df

allocated = table.groupby("date").apply(allocate)
allocated
```

```python
first = allocated.loc["2018-01-01", ["allocation"]]
last  = allocated.loc["2018-05-19", ["allocation"]]

merged = pd.concat([first, last], axis=1).fillna(0)
merged.columns = ["1/1/2018", "5/19/2018"]

merged.plot.bar(title="Top 20 Composition by Index Allocation: 1/1/18 x 5/19/18", figsize=(18, 8))

plt.ylabel("market cap (billions)")
plt.xlabel("symbol")

plt.show()
```

```python
top_4 = table.loc[(slice(None), ["BTC", "ETH", "XRP", "BCH"]), :].swaplevel(0,1).sort_index()

def calculate_performance(df):
    df["performance"] = (df["close"].pct_change() + 1).cumprod()
    return df

top_4 = top_4.groupby("symbol").apply(calculate_performance)
top_4 = top_4[["performance"]].unstack(level=0)
top_4 = top_4.set_index(pd.to_datetime(top_4.index))
top_4 = top_4[["performance"]]
top_4
```

```python
top_4.plot(title="Performance - HODL Only BTC, ETH, XRP or BCH: 1/1/18 - 5/19/18", figsize=(18, 8))
```
